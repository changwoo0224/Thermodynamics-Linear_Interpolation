```python
import numpy as np
import pandas as pd
import sys
Table = pd.read_excel('A_17.xlsx') # excel 파일 불러오기
print("A_17 Table 선형 보간 프로그램 by HCW")
print("Beta Ver")
A = input("온도: T, 엔탈피: h, 상대압력: Pr, 내부에너지: u, 상대 비체적:vr, 엔트로피:s를 입력하세요. 대문자 소문자 구분!\n")
['T','h','Pr','s','u','vr']
if A not in ['T','h','Pr','s','u','vr']:
    print("값을 잘 못 입력했습니다.")
    sys.exit()


X = float(input("기준값을 입력하세요\n")) # 알고 있는 값


B = input("구하고 싶은 상태량을 입력하세요. \n 온도: T, 엔탈피: h, 상대압력: Pr, 내부에너지: u, 상대 비체적:vr, 엔트로피:s를 입력하세요. 대문자 소문자 구분!\n ")
if B not in ['T','h','Pr','s','u','vr']:
    print("값을 잘 못 입력했습니다.")
    sys.exit()


# data table에서 필요한 값 추출 하는 함수


def extract_values(series, target_value):
    # 값이 테이블안에 있을 경우
    if target_value in series.values:
        extracted_values = target_value
    else:
        smaller_values = series[series < target_value].tail(1)  # target_value 보다 작은 값 중 가장 큰 값
        larger_values = series[series > target_value].head(1)  # target_value 보다 큰 값 중 가장 작은 값
        extracted_values = smaller_values.append(larger_values)
        
    return extracted_values



DataX = extract_values(Table[A],X) # 입력된 값 전 후 값 추출

# 입력된 값이 Table에 없을경우
if str(type(DataX)) == "<class 'pandas.core.series.Series'>":

    print("--------------")
    print("x값 맞는지 확인용")
    print(DataX)
    print("--------------\n")
    # 시리즈에서 x값 나눠주기
    x1 = float(DataX.iloc[0])  # 첫 번째 값
    x2 = float(DataX.iloc[1])  # 두 번째 값

    # x값의 행 값 찾기
    xrow1 = Table.index[(Table[A]==x1)] # 첫번째 값의 시리즈에서의 행 값
    xrow2 = Table.index[(Table[A]==x2)] # 두번째 값의 시리즈에서의 행 값

    # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
    y1 = float(Table[B].iloc[xrow1])
    y2 = float(Table[B].iloc[xrow2])

    
    delta = round((y2-y1)/(x2-x1),6) # 기울기 

    Total = round(delta*(X-x1)+y1,6)

    print(B+" 값은"+str(Total)+"입니다.\n")
    print("결과값은 소숫점 6째 자리 까지 나옵니다.")
# 입력된 값이 Table에 있을 경우    
else:
    
    x1 = DataX
    xrow = Table.index[(Table[A]==x1)] # x값의 행 값 찾아주기
    y1 = float(Table[B].iloc[xrow])
    
    print(B+" 값은 "+str(y1)+"입니다.\n")
    
```
