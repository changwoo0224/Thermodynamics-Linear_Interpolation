```python
print("A-4 Table 선형 보간 프로그램, 건도용 by HCW")
print("Beta Ver")

def extract_values(series, target_value):
    if target_value in series.values:
        extracted_values = target_value
        
    else:
        smaller_values = series[series < target_value].tail(1)  # target_value 보다 작은 값 중 가장 큰 값
        larger_values = series[series > target_value].head(1)  # target_value 보다 큰 값 중 가장 작은 값
        extracted_values = smaller_values.append(larger_values)
        
    return extracted_values

# 선형 보간 함수 x1,y1,x2,y2에는 이미 있는 값을, A는 알고 있는 값을 입력하시오.
def lin_interpolation(x1,y1,x2,y2,A):
    lin_data=round(((y2-y1)/(x2-x1))*(A-x1)+y1,6)
    
    return lin_data


A = input("T, Psat, vf,vg,uf,ufg,ug,hf,hfg,hg,sf,sfg,sg를 입력하세요. 대문자 소문자 구분!\n")


if A not in ['T', 'Psat', 'vf','vg','uf','ufg','ug','hf','hfg','hg','sf','sfg','sg']:
    print("값을 잘 못 입력했습니다.")
    sys.exit()

    
Wanttoknow= input("건도를 알고 싶으면 G, 건도로 u,h,s,v를 구하고 싶으면 u,h,s중 아무거나 입력하시오. ")


if Wanttoknow not in ['u','h','s','v','G']:
        print("값을 잘 못 입력했습니다.")
        sys.exit()
        
        
if Wanttoknow in ['u','h','s','v']:
    
    X=float(input("처음에 입력한 상태량의 값을 입력하세요\n")) # 알고 있는 값
    m=float(input("건도를 입력하세요\n")) # 알고 있는 값
    DataX = extract_values(Table4[A],X) # 입력된 값 전 후 값 추출
    if str(type(DataX))=="<class 'pandas.core.series.Series'>":
        print("--------------")
        print("x값 맞는지 확인용")
        print(DataX)
        print("--------------")
        # 시리즈에서 x값 나눠주기
    
        x1 = float(DataX.iloc[0])  # 첫 번째 값
        x2 = float(DataX.iloc[1])  # 두 번째 값
        xrow1 = Table4.index[(Table4[A]==x1)] # 첫번째 값의 시리즈에서의 행 값
        xrow2 = Table4.index[(Table4[A]==x2)] # 두번째 값의 시리즈에서의 행 값
        # 내부 에너지
        if Wanttoknow == 'u':
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['uf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['uf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['ug'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['ug'].iloc[xrow2]) # 두번째 ug
            ufg1 = float(Table4['ufg'].iloc[xrow1]) # 첫번째 ufg
            ufg2 = float(Table4['ufg'].iloc[xrow2]) # 두번째 ufg
            # 선형보간
            uf_lin = lin_interpolation(x1,y1,x2,y2,X)
            ug_lin = lin_interpolation(x1,y3,x2,y4,X)
            ufg_lin = lin_interpolation(x1,ufg1,x2,ufg2,X)
            totu=round(uf_lin+(ufg_lin)*m,6)
            print(totu)
        # 엔탈피    
        elif Wanttoknow == 'h':
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['hf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['hf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['hg'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['hg'].iloc[xrow2]) # 두번째 ug
            hfg1 = float(Table4['hfg'].iloc[xrow1]) # 첫번째 ufg
            hfg2 = float(Table4['hfg'].iloc[xrow2]) # 두번째 ufg
            # 선형보간
            hf_lin = lin_interpolation(x1,y1,x2,y2,X)
            hg_lin = lin_interpolation(x1,y3,x2,y4,X)
            hfg_lin = lin_interpolation(x1,hfg1,x2,hfg2,X)
            toth=round(hf_lin+(hfg_lin)*m,6)
            print(toth)
        # 비체적    
        elif Wanttoknow == 'v':
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['vf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['vf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['vg'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['vg'].iloc[xrow2]) # 두번째 ug
            # 선형보간
            vf_lin = lin_interpolation(x1,y1,x2,y2,X)
            vg_lin = lin_interpolation(x1,y3,x2,y4,X)
            totv = round(vf_lin+(vg_lin-vf_lin)*m,6)
            print(totv)
        # 엔트로피    
        else:
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['sf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['sf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['sg'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['sg'].iloc[xrow2]) # 두번째 ug
            sfg1 = float(Table4['sfg'].iloc[xrow1]) # 첫번째 ufg
            sfg2 = float(Table4['sfg'].iloc[xrow2]) # 두번째 ufg
            #선형보간
            sf_lin = lin_interpolation(x1,y1,x2,y2,X)
            sg_lin = lin_interpolation(x1,y3,x2,y4,X)
            sfg_lin = lin_interpolation(x1,sfg1,x2,sfg2,X)
            tots = round(sf_lin+(sfg_lin)*m,6)
            print(tots)
    # 테이블에 값이 있을경우        
    else:
        xrow = Table4.index[(Table4[A]==DataX)] # 첫번째 값의 시리즈에서의 행 값
        if Wanttoknow == 'u':
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['uf'].iloc[xrow]) # 해당하는 uf
            fg1 = float(Table4['ufg'].iloc[xrow]) # 해당하는 ufg
            y2 = float(Table4['ug'].iloc[xrow]) # 해당하는 ufg
            totu=round(y1+(fg1)*m,6)
            print(totu)
        
        elif Wanttoknow == 'h':
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['hf'].iloc[xrow]) # 해당하는 hf
            fg1 = float(Table4['hfg'].iloc[xrow]) # 해당하는 hfg
            y2 = float(Table4['hg'].iloc[xrow]) # 해당하는 hfg
            toth=round(y1+(fg1)*m,6)
            print(toth)
            
        elif Wanttoknow == 'v':
            # x값에서 구한 행 값으로 원하는 상태량의 y1, y2값 찾기
            y1 = float(Table4['vf'].iloc[xrow]) # 해당하는 vf
            y2 = float(Table4['hg'].iloc[xrow]) # 해당하는 vg
            totv=round(y1+(y2-y1)*m,6)
            print(totv)
            
        else:
            y1 = float(Table4['sf'].iloc[xrow]) # 해당하는 uf
            fg1 = float(Table4['sfg'].iloc[xrow]) # 해당하는 ufg
            y2 = float(Table4['sg'].iloc[xrow]) # 해당하는 ufg
            totu=round(y1+(fg1)*m,6)
            print(tots)
else:
    X=float(input("처음에 입력한 상태량의 값을 입력하세요\n")) # 알고 있는 값
    state = input("알고 있는 상태량을 입력하세요.(내부에너지:u,엔탈피:h, 엔트로피:s,대문자 소문자 구분!\n ")
    if state not in ['u','s','v','h']:
        print("잘못 입력 했습니다")
        sys.exit()
    state_data = float(input("알고 있는 상태량의 데이터를 입력하세요\n"))
    DataX = extract_values(Table4[A],X) # 입력된 값 전 후 값 추출 or 정확한 값 추출
    
    if str(type(DataX))=="<class 'pandas.core.series.Series'>":
        # 시리즈에서 x값 나눠주기
        x1 = float(DataX.iloc[0])  # 첫 번째 값
        x2 = float(DataX.iloc[1])  # 두 번째 값
        xrow1 = Table4.index[(Table4[A]==x1)] # 첫번째 값의 시리즈에서의 행 값
        xrow2 = Table4.index[(Table4[A]==x2)] # 두번째 값의 시리즈에서의 행 값
        
        if state == 'u':
            y1 = float(Table4['uf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['uf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['ug'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['ug'].iloc[xrow2]) # 두번째 ug
            ufg1 = float(Table4['ufg'].iloc[xrow1]) # 첫번째 ufg
            ufg2 = float(Table4['ufg'].iloc[xrow2]) # 두번째 ufg
            # 선형보간
            uf_lin = lin_interpolation(x1,y1,x2,y2,X)
            ug_lin = lin_interpolation(x1,y3,x2,y4,X)
            ufg_lin = lin_interpolation(x1,ufg1,x2,ufg2,X)
            gundo=(state_data-y1)/fg1
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
            
        elif state == 'h':
            y1 = float(Table4['hf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['hf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['hg'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['hg'].iloc[xrow2]) # 두번째 ug
            hfg1 = float(Table4['hfg'].iloc[xrow1]) # 첫번째 ufg
            ufg2 = float(Table4['hfg'].iloc[xrow2]) # 두번째 ufg
            # 선형보간
            hf_lin = lin_interpolation(x1,y1,x2,y2,X)
            hg_lin = lin_interpolation(x1,y3,x2,y4,X)
            hfg_lin = lin_interpolation(x1,hfg1,x2,ufg2,X)
            gundo=(state_data-y1)/fg1
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
        elif state == 's':
            y1 = float(Table4['sf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['sf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['sg'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['sg'].iloc[xrow2]) # 두번째 ug
            ufg1 = float(Table4['sfg'].iloc[xrow1]) # 첫번째 ufg
            ufg2 = float(Table4['sfg'].iloc[xrow2]) # 두번째 ufg
            # 선형보간
            sf_lin = lin_interpolation(x1,y1,x2,y2,X)
            sg_lin = lin_interpolation(x1,y3,x2,y4,X)
            sfg_lin = lin_interpolation(x1,sfg1,x2,sfg2,X)
            gundo=(state_data-y1)/fg1
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
        else:
            y1 = float(Table4['vf'].iloc[xrow1]) # 첫번째 uf
            y2 = float(Table4['vf'].iloc[xrow2]) # 두번째 uf
            y3 = float(Table4['vg'].iloc[xrow1]) # 첫번째 ug
            y4 = float(Table4['vg'].iloc[xrow2]) # 두번째 ug
            # 선형보간
            vf_lin = lin_interpolation(x1,y1,x2,y2,X)
            vg_lin = lin_interpolation(x1,y3,x2,y4,X)
            gundo = round(((state_data-vf)/(vg_lin-vf_lin)),6)
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
    else:
        xrow = Table4.index[(Table4[A]==DataX)] # 첫번째 값의 시리즈에서의 행 값
            
        if state == 'u':
            y1 = float(Table4['uf'].iloc[xrow]) # 해당하는 uf
            fg1 = float(Table4['ufg'].iloc[xrow]) # 해당하는 ufg
            y2 = float(Table4['ug'].iloc[xrow]) # 해당하는 ug
            gundo=(state_data-y1)/fg1
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
            
        elif state == 'h':
            y1 = float(Table4['hf'].iloc[xrow]) # 해당하는 hf
            fg1 = float(Table4['hfg'].iloc[xrow]) # 해당하는 hfg
            y2 = float(Table4['hg'].iloc[xrow]) # 해당하는 hg
            gundo=(state_data-y1)/fg1
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
        elif state == 's':
            y1 = float(Table4['sf'].iloc[xrow]) # 해당하는 sf
            fg1 = float(Table4['sfg'].iloc[xrow]) # 해당하는 sfg
            y2 = float(Table4['sg'].iloc[xrow]) # 해당하는 sg
            gundo=(state_data-y1)/fg1
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
        else:
            y1 = float(Table4['vf'].iloc[xrow]) # 해당하는 vf
            y2 = float(Table4['ug'].iloc[xrow]) # 해당하는 vg
            gundo = (state_data-v1)/(y2-y1)
            if gundo > 1.1:
                print(gundo)
                print("건도가 1보다 클 수 있을까? 다시 생각해보렴")
            else:
                print(gundo)
```
