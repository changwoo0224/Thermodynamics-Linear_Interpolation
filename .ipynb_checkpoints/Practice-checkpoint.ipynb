```python
import pandas as pd

def check_value_in_series(series, value):
    return value in series.values

# 시리즈 생성
s = pd.Series([1, 2, 3, 4, 5])

# 값을 확인하고 싶은 값
target_value = 3

# 값이 시리즈 안에 있는지 확인
result = check_value_in_series(s, target_value)

# 결과 출력
print(result)
```
